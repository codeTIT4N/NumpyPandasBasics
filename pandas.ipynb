```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#In case of numpy we were essentially creating n dimentional arrays but in case of pandas the core data structure is
#series
#To create series
x=pd.Series([1,2,3,4,5]) #passing a list in it
print(x)
```

```python
#Why series? Because we can do a bunch of different operations if we represent list as a series
#say you have to add 100 to all of the elements we only have to do
print (x+100)
```

```python
#we can also check which are the numbers greater than 2 for eg
print(x)
print(x>2)
```

```python
#any function
larger_than_2 = x > 2
print(larger_than_2)
larger_than_2.any()
#here any() will search in the whole Series if any of the value is true it will return true else false
```

```python
#all function
print(larger_than_2)
larger_than_2.all()
#similarly here all() will search in the whole Series if any of the value is true it will return true else false
```

```python
#apply
#Say you have a list and to each of the element of the list you have to apply a function 
#for example you need to square each of the elements, etc .i.e., you have to perform complex operation that 
#can't be done using these list comprehentions then it has apply which does it efficiently

#to do that first write a function 
def f(x):
    if x%2 ==0:
        return x**2   #remember ** is ^ in python
    else:
        return x**3
#so this function is squaring all the even values of x and cubing all the odd values of x

#now use apply()
print(x.apply(f))
#this apply the function to all the elements in the series
```

```python
#converting types use astype()
x.astype(np.float64)
#here we have to pass the type so we use numpy reference here to pass float64
```

```python
#copy
y=x 
#this is not a normal copy if we change anything in y it will automatically get changed in x too
y[0]=100
print(x)
print(y)
```

```python
#so to create a proper copy of a series object and store it in another variable we have to use copy function

#first reinitializing x
x=pd.Series([1,2,3,4,5])

#now using copy function
y=x.copy()

#now changing a certain value in y
y[2]=200
print(x)
print(y)
#this is called a proper copy
```

```python
#Dataframes : it is the main datastructure of pandas
#A dataframe is like a list of series example:
data = [1,2,3,4,5,6,7,8,9] #normal python list

#now creating a dataframe
df=pd.DataFrame(data,columns=["x"]) #x would be the name of the column
print(df)
#
```

```python
#to get series out of a dataframe
dataSeries = df["x"]
print(dataSeries)
#now as we can see they look the same but the difference between them is that a dataframe can have multiple columns
#only in this particular case we made just one column but there could be many columns in a dataframe 
#so essentially it is like a bunch of series stacked together whereas series contains a single list
```

```python
#Adding more columns in dataframe
df["x_plus_2"] = df["x"]+2
print(df)
```

```python
#similarly adding a square column
df["x_square"] = df["x"]**2
print(df)
```

```python
#similarly adding a factorial column
df["x_factorial"] = df["x"].apply(np.math.factorial)  #note here we are passing np's math.factorial function
print(df)
```

```python
#similarly adding a is even column
df["is_even"] = df["x"]%2
df #note we can also print df like this IMPORTANT
```

```python
#deleting a column use drop function
df = df.drop("is_even",1) #this deletes the is_even column and 1 means how many columns are we deleting starting from
                          #is_even column
df
```

```python
#till now we are only displaying the whole dataframe but what if we want to display specific columns
#Selecting multiple columns 
df[["x","x_square"]]
#note here we are passing a list of columns that we want to display
```

```python
#describing
df.describe()
#this will give you all the different information about the dataframe like count, mean, std(standard deviation),etc
```
